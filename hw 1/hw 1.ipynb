```python
from pymorphy2 import MorphAnalyzer
morph = MorphAnalyzer()
```

```python
a = t.replace('\n', '').lower()
```

```python
import re
```

```python
def punct(text): #function to get rid of punctuation
    te = [re.sub('[^\w\s]','', text)]
    return te
```

```python
j = punct(a)
```

```python
words = j[0].split()
```

```python
d = dict()
```

```python
#pip install jsonlines
```

```python
import jsonlines
```

```python
for w in words: #jsonlines with pymorphy analysis
    ana = morph.parse(w)
    lemma = ana[0].normal_form
    word = ana[0].word
    pos = ana[0].tag.POS
    d['lemma'] = lemma
    d['word'] = word
    d['pos'] = pos
    with jsonlines.open('output.jsonl', mode='a') as writer:
        writer.write(d)
```

```python
from collections import Counter
```

```python
num = Counter()
```

```python
co = 0 #counter for number of words (=lines in jsonl)
```

```python
with jsonlines.open('output.jsonl') as reader: #counting different parts of speech
    for line in reader.iter(type=dict, skip_invalid=True):
        pos = line['pos']
        num[pos] += 1
        co = co + 1
```

```python
num
```

```python
co #total number of words
```

```python
for i in num:
    proc = num.get(i)/co*100
    num[i] = proc
```

```python
num #проценты каждой части речи
```

```python
num_v = Counter() #counters for verbs and adverbs
num_adv = Counter()
```

```python
with jsonlines.open('output.jsonl') as reader: #counting verbs and adverbs
    for line in reader.iter(type=dict, skip_invalid=True):
        lemma = line['lemma']
        if line['pos'] == 'VERB':
            num_v[lemma] += 1
        elif line['pos'] == 'ADVB':
            num_adv[lemma] += 1
```

```python
num_v.most_common(20) #top-20 verbs
```

```python
num_adv.most_common(20) #top-20 adverbs
```

```python
sent = t.replace('\n', '') #choosing several sentences from the text
sent = '.'.join(sent.split('.')[200:205])
s_words = sent.split()
s_words
```

```python
full_s = []
```

```python
for w in s_words: #changing inflection
    ana = morph.parse(w)
    if ana[0].tag.POS == 'VERB':
        full_s.append(ana[0].inflect({'past', 'masc'}).word)
    elif ana[0].tag.POS == 'NOUN':
        full_s.append(ana[0].inflect({'plur', 'gent'}).word)
    else:
        full_s.append(ana[0].word)
```
