```python
m = 'model.bin'

if m.endswith('.vec'):
    model = gensim.models.KeyedVectors.load_word2vec_format(m, binary=False)
elif m.endswith('.bin'):
    model = gensim.models.KeyedVectors.load_word2vec_format(m, binary=True)
else:
    model = gensim.models.KeyedVectors.load(m)
```

```python
words = ['бегать_VERB', 'скакать_VERB', 'прыгать_VERB','идти_VERB', 'бежать_VERB', 'ползти_VERB', 'ползать_VERB', 'пройти_VERB', 'перейти_VERB', 'пролезть_VERB', 'плыть_VERB', 'ходить_VERB', 'убегать_VERB', 'убежать_VERB', 'прыжок_NOUN', 'скачок_NOUN', 'пробежка_NOUN', 'побег_NOUN', 'ходьба_NOUN', 'плавание_NOUN']
```

```python
X = model[words]
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
coords = pca.fit_transform(X)
```

```python
df_verbs = pd.DataFrame(coords, columns = ['x','y'])#dataframe from word2vec coordinates
```

```python
pos = []
lex = []
```

```python
for word in words:
    ann = word.split('_')
    lex.append(ann[0])
    pos.append(ann[1])
```

```python
df_verbs['lexeme'] = lex #column with lexeme
df_verbs['POS'] = pos #column with part of speech tag
df_verbs.head()
```

```python
N = 1000
fig = go.Figure(data=go.Scattergl(
   x = df_verbs['x'],
   y = df_verbs['y'],
   mode='markers+text',
   marker=dict(
       color=np.random.randn(N),
       size=15,
       symbol="diamond",
       line_width=2
   ), 
   text=df_verbs['lexeme'],
   textposition='top center',
   hovertext=df_verbs['POS']
))

fig.show()
```
